# IF ELSE

```python
salary=int(input("Enter the salary"))
if salary >= 70000:
  print("High Salary")
elif salary>= 50000:
  print("Average Salary")
else:
  print("Low Salary")
```

# Functions

```python
def tax_bracket(salary):
  if salary >= 70000:
    return "30%"
  elif salary>= 50000:
    return "20%"
  else:
    return "10%"
tax_bracket(100000)
```

```python
salary = int(input("Enter the salary"))
print("Tax Bracket is",tax_bracket(salary))
```
