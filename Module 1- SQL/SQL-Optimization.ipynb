```sql
select emp_id, emp_name, sale_amount, dept, 
rank() over(partition by dept order by sale_amount desc) as rank
from employee_sales;
```

```sql
select emp_name, sum(sale_amount) over(partition by emp_name order by sale_amount desc) as total_sales
from employee_sales;
```

```sql
SELECT emp_name,
sum(sale_amount) as sale_amt
FROM employee_sales
where sale_date > '2023-10-01'
GROUP by emp_name
```
